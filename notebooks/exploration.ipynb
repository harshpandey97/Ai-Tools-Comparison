# AI Tools Comparison: ChatGPT vs Claude vs Gemini

Interactive exploration notebook. Loads the same datasets used in `src/analyze.py` for ad-hoc analysis.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

bench = pd.read_csv('../data/ai_tools_benchmarks.csv')
survey = pd.read_csv('../data/user_survey.csv')
bench.head()
```

## 1. Benchmark scores by category

```python
perf = bench[bench['category'].isin(['reasoning','coding','math','multimodal'])]
pivot = perf.pivot_table(index='benchmark', columns='tool', values='score')
pivot
```

```python
pivot.plot(kind='bar', figsize=(9,5))
plt.title('Benchmark Scores by Tool')
plt.ylabel('Score (%)')
plt.tight_layout()
plt.show()
```

## 2. Which tool wins each benchmark?

```python
winners = pivot.idxmax(axis=1)
winners
```

## 3. Cost-efficiency: score per dollar (using SWE-bench + output pricing)

```python
swe = bench[bench['benchmark']=='SWE-bench'].set_index('tool')['score']
price = bench[bench['benchmark']=='output_cost_per_mtok'].set_index('tool')['score']
efficiency = (swe / price).sort_values(ascending=False)
efficiency
```

## 4. User survey: satisfaction vs accuracy correlation

```python
survey[['satisfaction_score','ease_of_use','accuracy_rating','monthly_usage_hours']].corr()
```

```python
sns.scatterplot(data=survey, x='accuracy_rating', y='satisfaction_score', hue='tool_primary')
plt.title('Satisfaction vs Accuracy Rating')
plt.show()
```
